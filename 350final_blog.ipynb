# Method

To compare the accuracy of Google translate and Amazon translate for formal vs informal language, and for latin root vs non-latin root language, the following metrics are chosen for the reasons listed below. 

## Qualitative comparison

**Metric 1**: Ratings of translated text from sample users 

Our group members will be the sample users that rate the texts that were translated into English from Chinese, Spanish, and French. In rating, we will compare the translated text to the official text and make a rating between 0 and 10.


## Quantitative comparison


**Metric 1**: Percentage of words that are accurate when comparing the translated text to the “gold standard official” translation


In terms of [Machine Translation](https://towardsdatascience.com/bleu-bilingual-evaluation-understudy-2b4eab9bcfd1), 

$Precision = \frac{number\,of \,candidate\, translation\, words\, occuring\, in\, any\, reference\, translation}{total\, number\,of\, words\, in\,candidate\, translation}$

Where candidate translation refers to the machine translation of interest, and reference translation refers to human reference translation that is established to fully convey the meaning of a text. In short, the more the number of matches between machine translation and human translation, the better the machine translation is. 

Therefore, our first metric is the percentage of words that overlap in the original text and translated text. However, this is a relatively low level metric as does not capture more complicated aspects of language such as the order of the overlapping words or the use of synonyms. Moreover, an [n-gram](https://devopedia.org/n-gram-model) refers to a sequence of words occurring together within a window of n words, which is important in producing a more natural sentence. Therefore, we used a more well established and well known metric, BLEU score.

**Metric 2**: [BLEU Score](https://towardsdatascience.com/nlp-metrics-made-simple-the-bleu-score-b06b14fbdbc1) stands for Bilingual evaluation Understudy. It is a [well established](https://machinelearninginterview.com/topics/natural-language-processing/bleu-score/#:~:text=What%20is%20BLEU%20Score%3F,manual%20evaluator%20or%20a%20translator.) metric used to "evaluate the quality of machine generated text by comparing it with a reference text that is supposed to be generated." Unlike metric 1, BLEU score calculation takes into account n-grams, how words are repeated, and many more. 

We consulted a [Bleu demo on python](https://colab.research.google.com/gist/bshmueli/ce62e435bbcb5b41f797c0b8c4d0f4b9/bleu-demo.ipynb) to obtain a more nuanced analysis of the translations.

# Results

We believe that out of metrics used, the Average Qualitative Scores and Average BLEU Scores are the most important. This is because the "Percentage of Qualtitative Scores above 7" doesn't distinguish between the groups as much, and the "Average Percent Words Match" score isn't as reliable because synonyms in two different translations may convey the same meaning even though the words themselves are different. 

Therefore, we think that the Average Qualitative Scores and Average BLEU Scores are the most important. The Qualitative Scores reflect how a person fluent in English would interpret the texts, and the BLEU score reflects a commonly-used, reliable NLP metric used to compare translations of a text. 

Overall, with the phrases chosen from the Great Gatsby, Google performed better with the Qualitative Analysis, BLEU score, and Match score. For the informal translations, AWS performed better. For the formal translations, Google performed better.

For Google and AWS, we see that as the Qualitative Score increased, so did the BLEU Score. Similarly, as the Qualitative Score increased, so did the Match Score. For both AWS and Google, the regression line for the BLEU score vs Qualitative Score is steeper, indicating a potentially stronger relationship. 

# Discussion


From the results with excerpts from the Great Gatsby, it can be inferred that Google can be used for more formal translations and therefore may be better for students who are learning a new language since classes typically teach languages in a more formal manner. Yet, AWS may be better when learning the more “informal” version of a language.


However, there are a few areas of improvement to our metrics and our data collection. First and foremost, quantitative metrics 1 nor 2 cannot handle semantic similarity. They both look for exact matches, but the quality of synonym use is an important aspect to incorporate in the future. 

Moreover, Neither Google translate nor Amazon translate was tested with "slangs" or modern internet languages. It is important to acknowledge that language continue to evolve. Therefore, a possible research in the future can be done with texts spanning between a long time frame.

Furthermore, it would be helpful if we could generate a distribution of each metric to see if there are statistically significant differences to make conclusive statements. The size of the data we have as of now is small, but with a sizeable amount of data in the future, we could verify the significance of each metric. 

Evaluating language translation is a tricky area since there is a high degree of subjectivity and nuance that cannot be translated into machine languages. Moreover, language continues to evolve, so investigation should  be done with a more diverse and bigger dataset and be done continuosly over time.
