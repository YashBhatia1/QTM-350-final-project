# Comparing AWS Translate and Google Translate

Importing Packages

```python
from statistics import mean
import matplotlib.pyplot as plt
import numpy as np
```

```python
qualitative_scores_aws = [8.5, 8.5, 9, 8.5, 8.5, 8.5, 3, 7, 7, 5, 9, 5]
qualitative_scores_google = [8.5, 9, 8, 8.5, 8.5, 8, 6.5, 7, 7, 6, 9, 5]
```

### Percentage of scores above 7 (AWS)

```python
good = 0
for i in qualitative_scores_aws:
    if i > 7:
        good +=1
        
percent_good = good * 100 / len(qualitative_scores_aws)
print(percent_good)
```

### Percentage of scores above 7 (Google)

```python
good = 0
for i in qualitative_scores_google:
    if i > 7:
        good +=1
        
percent_good = good * 100 / len(qualitative_scores_google)
print(percent_good)
```

Using Percentage of Qualitative Scores above 7 as a metric, there is no difference between AWS and Google Translate.

### Average Qualitative Scores (AWS)

```python
qualitative_scores_aws = [8.5, 8.5, 9, 8.5, 8.5, 8.5, 3, 7, 7, 5, 9, 5]
print(mean(qualitative_scores_aws))
```

### Average Qualitative Scores (Google)

```python
qualitative_scores_google = [8.5, 9, 8, 8.5, 8.5, 8, 6.5, 7, 7, 6, 9, 5]
print(mean(qualitative_scores_google))
```

Using Average Qualitative Scores as a metric, Google Translate performs slightly better than AWS Translate.

### Average BLEU Score (AWS)

```python
bleu_scores_aws = [0.0887, 0.3059, 0.07713, 0.2354, 0.2541, 0.4659, 1.9771549747291336e-78, 0.21416, 0.3014, 1.2627076138080564e-231, 4.0622028886850106e-78, 3.191385619211647e-78]
print(mean(bleu_scores_aws))
```

### Average BLEU Score (Google)

```python
bleu_scores_google = [0.1063, 0.33073, 0.08016, 0.3584, 0.3158, 0.52076, 0.24696, 0.17999, 0.4904, 1.2627076138080564e-231, 0.4408, 4.994788421695789e-78]
print(mean(bleu_scores_google))
```

Using Average BLEU Scores as a metric, Google Translate performs significantly better than AWS Translate.

### Average Percent Words Match (AWS)

```python
match_score_aws = [0.4043, 0.6191, 0.4324, 0.67442, 0.63415, 0.80435, 0.46154, 0.5, 0.5, 0.23077, 0.5833, 0.4444]
print(mean(match_score_aws))
```

### Average Percent Words Match (Google)

```python
match_score_google = [0.4634, 0.6136, 0.42105, 0.6444, 0.75, 0.7660, 0.6389, 0.4634, 0.7097, 0.25, 0.75, 0.6]
print(mean(match_score_google))
```

Using Average Percent Match Scores as a metric, Google Translate performs better than AWS Translate.

### Visualisations

```python
plt.plot(qualitative_scores_aws, color = "r", label = "AWS")
plt.plot(qualitative_scores_google, color = "b", label = "Google")
plt.xlabel('Translation')
plt.ylabel('Qualitative Score')
plt.legend()
plt.title("Comparison of AWS Translate and Google Translate using Qualitative Scores")
plt.show()

plt.plot(bleu_scores_aws, color = "r", label = "AWS")
plt.plot(bleu_scores_google, color = "b", label = "Google")
plt.xlabel('Translation')
plt.ylabel('BLEU Score')
plt.legend()
plt.title("Comparison of AWS Translate and Google Translate using BLEU Scores")
plt.show()

plt.plot(match_score_aws, color = "r", label = "AWS")
plt.plot(match_score_google, color = "b", label = "Google")
plt.xlabel('Translation')
plt.ylabel('Match Score')
plt.legend()
plt.title("Comparison of AWS Translate and Google Translate using Percent Match Scores")
plt.show()
```

# How well does our Qualitative Score match with the BLEU and Percent Match Scores?

### AWS

```python
qualitative_scores_aws = np.array(qualitative_scores_aws)
bleu_scores_aws = np.array(bleu_scores_aws)

plt.scatter(qualitative_scores_aws, bleu_scores_aws, color = "r")

m, b = np.polyfit(qualitative_scores_aws, bleu_scores_aws, 1)
plt.plot(qualitative_scores_aws, m*qualitative_scores_aws+b, label = "Regression Line")
plt.legend()
plt.xlabel("Qualitative Score")
plt.ylabel("BLEU Score")
plt.title("BLEU Score vs Qualitative Score - AWS")
plt.show()


match_score_aws = np.array(match_score_aws)

plt.scatter(qualitative_scores_aws, match_score_aws, color = "r")

m, b = np.polyfit(qualitative_scores_aws, match_score_aws, 1)
plt.plot(qualitative_scores_aws, m*qualitative_scores_aws+b, label = "Regression Line")
plt.legend()
plt.xlabel("Qualitative Score")
plt.ylabel("Match Score")
plt.title("Match Score vs Qualitative Score - AWS")
plt.show()
```

### Google

```python
qualitative_scores_google = np.array(qualitative_scores_google)
bleu_scores_google = np.array(bleu_scores_google)

plt.scatter(qualitative_scores_google, bleu_scores_google, color = "r")

m, b = np.polyfit(qualitative_scores_google, bleu_scores_google, 1)
plt.plot(qualitative_scores_google, m*qualitative_scores_google+b, label = "Regression Line")
plt.legend()
plt.xlabel("Qualitative Score")
plt.ylabel("BLEU Score")
plt.title("BLEU Score vs Qualitative Score - Google")
plt.show()


match_score_google = np.array(match_score_google)

plt.scatter(qualitative_scores_google, match_score_google, color = "r")

m, b = np.polyfit(qualitative_scores_google, match_score_google, 1)
plt.plot(qualitative_scores_google, m*qualitative_scores_google+b, label = "Regression Line")
plt.legend()
plt.xlabel("Qualitative Score")
plt.ylabel("Match Score")
plt.title("Match Score vs Qualitative Score - Google")
plt.show()
```

# Summary of Analysis

We believe that out of the 4 metrics, the Average Qualitative Scores and Average BLEU Scores are the most important. This is because the "Percentage of Qualtitative Scores above 7" doesn't distinguish between the groups as much, and the "Average Percent Words Match" score isn't as reliable because synonyms in two different translations may convey the same meaning even though the words themselves are different. 

Therefore, we think that the Average Qualitative Scores and Average BLEU Scores are the most important. The Qualitative Scores reflect how a person fluent in English would interpret the texts, and the BLEU score reflects a commonly-used, reliable NLP metric used to compare translations of a text. 

Overall, Google performed better with the Qualitative Analysis, BLEU score, and Match score. For the informal translations, AWS performed better. For the formal translations, Google performed better.

For Google and AWS, we see that as the Qualitative Score increased, so did the BLEU Score. Similarly, as the Qualitative Score increased, so did the Match Score. For both AWS and Google, the regression line for the BLEU score vs Qualitative Score is steeper, indicating a potentially stronger relationship.
